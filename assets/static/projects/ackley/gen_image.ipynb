```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def ackley(x):
    a = 20
    b = 0.2
    c = 1
    return (
        -a * np.exp(-b * np.sqrt(np.sum(x**2) / len(x)))
        - np.exp(np.sum(np.cos(c * x)) / len(x))
        + a
        + np.e
    )
```

```python
step = 0.25
for i in range(10):
    bounds = np.array([[-32.768, 32.768], [-32.768, 32.768]])
    x = np.linspace(-100, 100, 20)
    y = np.linspace(-100, 100, 20)
    X, Y = np.meshgrid(x, y)

    arrow_c = np.array([101, 126, 212]) / 255.0
    fig, ax = plt.subplots(figsize=(10, 8))
    for x, y in zip(X.flatten(), Y.flatten()):
        x /= (1 + 1 * step * i)
        y /= (1 + 1 * step * i)

        ax.arrow(x, y, -x/10, -y/10,
                width=0.1, head_width=0.5, head_length=0.5,
                fc=arrow_c, ec=arrow_c, zorder=3)

    x = np.linspace(bounds[0, 0], bounds[0, 1], 300)
    y = np.linspace(bounds[1, 0], bounds[1, 1], 300)
    X, Y = np.meshgrid(x, y)
    Z = np.array([ackley(np.array([x, y])) for x, y in zip(X.flatten(), Y.flatten())])
    Z = Z.reshape(X.shape)

    ax.contour(X, Y, Z, levels=20, linewidths=2, colors="black", zorder=2)

    ax.add_patch(plt.Rectangle((-32.768-0.5, -32.768-0.5),
                                32.768*2+1, 32.768*2+1,
                                color="white", zorder=1))

    ax.set_facecolor("#ffffff")

    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(bounds[0, :])
    ax.set_ylim(bounds[1, :])
    ax.spines[['top', 'right', 'left', 'bottom']].set_visible(False)
    ax.set_aspect('equal', adjustable='box')

    plt.savefig(f"images/ackley_contour_{i}.png",
                bbox_inches="tight",
                pad_inches=0,
                dpi=300)
    plt.clf()
    print(f"Saved image {i}")
```

```python
from moviepy import ImageSequenceClip
from pathlib import Path
import re

def numeric_key(p: Path):
    m = re.search(r'(\d+)\.png$', p.name)
    return int(m.group(1)) if m else p.name

frames = sorted(Path("images").glob("ackley_contour_*.png"), key=numeric_key)
assert frames, "Aucune image trouvée dans images/"

# ...existing code...
clip = ImageSequenceClip([str(p) for p in frames], fps=4)
clip.write_videofile(
    "ackley_contour.webm",
    codec="libvpx",
    audio=False,
    fps=4,
    ffmpeg_params=["-crf", "38", "-b:v", "0"]  # qualité: CRF bas = meilleure qualité
)
```
